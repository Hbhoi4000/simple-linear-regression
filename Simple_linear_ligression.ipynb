```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("weight-height.csv")
df.drop("Gender",axis=1,inplace=True)
df = df.sample(n = 1000) 
```

```python
df.isnull().sum()
```

```python
plt.scatter(df["Weight"],df["Height"])
plt.xlabel("weight")
plt.ylabel("Height")
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
X=df[["Weight"]]   #independant variable should be in 2 d array or dataframe 
Y=df["Height"]    #  depabdnt variable should be in single array or serries 
```

```python
np.array(X).shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.75,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
x_train =scaler.fit_transform(X_train)
x_test =scaler.transform(X_test)
```

```python
print("x train",x_train.shape)
print("Y_train",y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(x_train,y_train)
```

```python
print("slope of the line is ",regression.coef_)
print("intercept of the line is ",regression.intercept_)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,regression.predict(x_train))
```

```python
y_predict=regression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python
print(" mean sqaure error:",mean_squared_error(y_test,y_predict))
print(" mean absolute error:",mean_absolute_error(y_test,y_predict))
print("Root mean sqaure error:",root_mean_squared_error(y_test,y_predict))
```
